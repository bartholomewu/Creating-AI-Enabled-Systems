```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('Musical_instruments_reviews.csv')
data
```

```python
summary = data['summary']
summary
```

```python
summary_l = summary.to_list()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk
```

```python
from nltk.stem import PorterStemmer
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
def tokenizer(data_l):
    tokens = []
    for data in data_l:
        tokens.append(nltk.word_tokenize(data))
    return tokens
```

```python
def stemmer(data_l):
    ps = PorterStemmer()
    stemmed = []
    for element in data_l:
        if ' ' in element:
            tmp = []
            el_sp = element.split()
            for el in el_sp:
                tmp.append(ps.stem(el))
            stemmed.append(' '.join(tmp))
        else:
            stemmed.append(ps.stem(element))
    return stemmed
```

```python
def lemmar(data_l):
    lmtzr = WordNetLemmatizer()
    lemma = []
    for element in data_l:
        if ' ' in element:
            tmp = []
            el_lm = element.split()
            for el in el_lm:
                tmp.append(lmtzr.lemmatize(el, 'v'))
            lemma.append(' '.join(tmp))
        else:
            lemma.append(lmtzr.lemmatize(element, 'v'))
    return lemma
```

```python
tokens = tokenizer(summary_l)
tokens
```

```python
stemmed = stemmer(summary_l)
stemmed
```

```python
lemmed = lemmar(summary_l)
lemmed
```
