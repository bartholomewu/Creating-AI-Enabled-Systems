```python
import numpy as np 
import matplotlib.pyplot as plt 
import cv2 
```

```python
images = ('BikeShot1.jpg', 'BikeShot2.jpg', 'BikeShot3.jpg', 'BikeShot4.jpg')
```

```python
def resize(images):
    img_resized = []
    for image in images:
        tmp = cv2.imread(image)
        dsize = (500, 500)
        img_resized.append(cv2.resize(tmp, dsize, interpolation=cv2.INTER_AREA))
        
    return img_resized
```

```python
def normalize(images):
    img_normalized = []
    dsize = (500, 500)
    normalized_img = np.zeros(dsize)
    for image in images:
        tmp = cv2.imread(image)
        img_normalized.append(cv2.normalize(tmp, normalized_img, 0, 255, cv2.NORM_MINMAX))
    return img_normalized
```
